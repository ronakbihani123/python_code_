```python
df = pd.read_csv("/content/covid_toy.csv")
df.head(2)
```

```python
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import OneHotEncoder
from sklearn.preprocessing import OrdinalEncoder
```

```python
df.isnull().sum()
```

```python
x = df.drop(columns=['has_covid'])
y = df['has_covid']
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train , x_test,y_train ,y_test = train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
x_train.head(2)
```

```python
# manually process
```

```python
# adding simple imputer to fever column
si = SimpleImputer(strategy='mean')
x_train_fever = si.fit_transform(x_train[['fever']])
x_train_fever.shape
```

```python
# OrdinalEncoder---> cough
oe = OrdinalEncoder(categories=[['Mild','Strong']])
x_train_cough = oe.fit_transform(x_train[['cough']])
x_train_cough.shape
```

```python
# onehotencoding --> gender , city
ohe = OneHotEncoder(drop='first',sparse_output=False)
x_train_gender_city = ohe.fit_transform(x_train[['gender','city']])
x_train_gender_city.shape
```

```python
# Extracting age
x_train_age = x_train.drop(columns=['gender','fever','cough','city']).values
x_train_age.shape
```

```python
x_train_transformed = np.concatenate((x_train_age,x_train_fever,x_train_gender_city,x_train_cough),axis=1)
# also the test data
```

```python
x_train_transformed.shape
```

```python
# we will do this same work with the help of column transformer
```

```python
from sklearn.compose import ColumnTransformer     #this is hoe to import column _transformer
```

```python
transformer = ColumnTransformer(transformers=[('tnf1',SimpleImputer(),['fever']),           #in a fever column by the help of si we fill missing values by mean,median,mode
                                               ('tnf2',OrdinalEncoder(categories=[['Mild','Strong']]),['cough']),      #by this process we encode our data
                                               ('tnf3',OneHotEncoder(drop='first',sparse_output=False),['gender','city'])],remainder='passthrough')
                                                    # remainder  =  passthrough ===>
                                                    # it means rest all the column remains same
```

```python
a = transformer.fit_transform(x_train)
```

```python
a.shape
```

```python
from sklearn.preprocessing import FunctionTransformer
```

```python
x = np.array([[1,2],[3,4]])   #create dataset
```

```python
def squ(x):
  return np.hstack((x,x**2))   #define a custom feature engineering function
```

```python
custom_transformer = FunctionTransformer(squ)         #create a function transformer to apply the custom function
```

```python
x_transformed  = custom_transformer.transform(x)       #apply the transformer to the input data
```

```python
print(x_transformed)               #view the transformed data
```
